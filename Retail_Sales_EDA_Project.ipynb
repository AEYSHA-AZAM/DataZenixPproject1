```python
import seaborn as sns
import matplotlib.pyplot as plt

# Monthly Sales Trend
monthly_sales = df.groupby('Month')['Sales'].sum().reset_index()
monthly_sales['Month'] = monthly_sales['Month'].astype(str)

# Top Products
top_products = df.groupby('Product')['Sales'].sum().sort_values(ascending=False).reset_index()

# Heatmap
pivot = df.pivot_table(index='Product', columns='Month', values='Sales', aggfunc='sum')

# Plotting
sns.set(style='whitegrid')
plt.rcParams['figure.figsize'] = (14, 18)
fig, axs = plt.subplots(3, 1, figsize=(14, 18))

# 1. Line Plot
sns.lineplot(data=monthly_sales, x='Month', y='Sales', marker='o', ax=axs[0])
axs[0].set_title("Monthly Sales Trend")
axs[0].tick_params(axis='x', rotation=45)

# 2. Bar Plot
sns.barplot(data=top_products.head(5), x='Sales', y='Product', palette='viridis', ax=axs[1])
axs[1].set_title("Top 5 Products by Sales")

# 3. Heatmap
sns.heatmap(pivot, cmap="YlGnBu", ax=axs[2], linecolor='white', linewidths=0.5)
axs[2].set_title("Heatmap of Product Sales per Month")

plt.tight_layout()
plt.show()
```
